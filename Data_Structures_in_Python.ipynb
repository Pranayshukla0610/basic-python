**DATA STRUCTURES IN PYTHON**

```python
#List – Remove Duplicates Without Using Set

def remove_duplicates(nums):
  result = []
  for n in nums:
    if n not in result:
      result.append(n)
  return result

print(remove_duplicates([1,2,2,3,4,4,5]))
```

```python
#List – Flatten a Nested List

def flatten_list(nested):
  flat = []

  for i in nested:
    if isinstance(i, list):
      flat.extend(i)
    else:
      flat.append(i)
  return flat

print(flatten_list([1, [2,3], [4,[5,6]]]))
```

```python
#Check if a variable is iterable but not a string
from collections.abc import Iterable

def is_iterable(obj):
  return isinstance(obj, Iterable) and not isinstance(obj, (str,bytes))

print(is_iterable([1,2,3,4,5]))
print(is_iterable('abc'))
```

```python
x = 3.14
print(isinstance(x,(int,float)))
```

```python
#Implement a function that accepts only numbers (int/float) and raises error otherwise
def safe_add(a,b):
  if not isinstance(a,(int,float)) or not isinstance(b,(int,float)):
    raise TypeError('Inputs must be int or float')
  return a + b

print(safe_add(2.5,5))
```

```python
#Polymorphic function that behaves differently by type

def describe(x):
  if isinstance(x,int):
    return f'{x} is an integer'
  elif isinstance(x,str):
    return f'{x} is an string'
  elif isinstance(x,list):
    return f'{x} is a list'
  else:
    return 'Unknown Type'

print(describe(42))
print(describe('Hello'))
print(describe([1,2,3]))
```
