```python
#Check File Information (size, type, validity)

import os
def check_file_info(file_path):
  if not os.path.exists(file_path):
    print("File does not exists")
    return

  size = os.path.getsize(file_path)

  print(f"File Path: {file_path}")
  print(f"File Size: {size/1024:2f}KB")
  print(f"File Size: {size/(1024*1024):.2f} MB")

check_file_info('/content/archive (1).zip')
```

```python
#Delete File Type Automatically

import mimetypes
def detect_file_type(file_path):
  file_type = mimetypes.guess_type(file_path)
  print("Detected Type:",file_type)

detect_file_type('/content/archive (1).zip')
```

```python
#Extract Zip File

import zipfile

def extract_zip(file_path, extract_to="extracted_zip"):
    if zipfile.is_zipfile(file_path):
        with zipfile.ZipFile(file_path, 'r') as zip_ref:
            zip_ref.extractall(extract_to)
        print("✅ ZIP extracted successfully")
    else:
        print("❌ Not a valid ZIP file")

extract_zip("/content/archive (1).zip")
```

```python
import zipfile

zip_path = "/content/archive (1).zip"

with zipfile.ZipFile(zip_path, 'r') as z:
    for file in z.namelist():
        try:
            z.extract(file, "data")
            print("Extracted:", file)
        except:
            print("Skipped corrupted file:", file)
```

```python
import os

def verify_download(path, min_size_mb=10):
    size = os.path.getsize(path)/(1024*1024)
    print("File size:", round(size,2),"MB")

    if size < min_size_mb:
        print("⚠️ Download likely incomplete")

verify_download("/content/archive (1).zip")
```

```python
import tarfile

def extract_tar(file_path, extract_to = 'extracted_tar'):
  try:
    with tarfile.open(file_path) as tar:
      tar.extractall(extract_to)
    print("✅ TAR extracted successfully")
  except:
    print("❌ Not a valid TAR file")

extract_tar("archive.tar.gz")
```

```python
import rarfile

def extract_rar(file_path, extract_to="extracted_rar"):
    try:
        with rarfile.RarFile(file_path) as rf:
            rf.extractall(extract_to)
        print("✅ RAR extracted successfully")
    except:
        print("❌ Not a valid RAR file")

extract_rar("archive.rar")
```

```python
def extract_7z(file_path, extract_to="extracted_7z"):
    try:
        with py7zr.SevenZipFile(file_path, mode='r') as z:
            z.extractall(path=extract_to)
        print("✅ 7z extracted successfully")
    except:
        print("❌ Not a valid 7z file")

extract_7z("archive.7z")
```

```python
import zipfile

def list_zip_contents(file_path):
    with zipfile.ZipFile(file_path, 'r') as zip_ref:
        print(zip_ref.namelist())

list_zip_contents("/content/archive (1).zip")
```

```python
import os

def count_files(folder_path):
    total_files = 0

    for root, dirs, files in os.walk(folder_path):
        total_files += len(files)

    print("📂 Total files:", total_files)

count_files("/content")
```

```python
import zipfile
import tarfile
import os

def extract_file(file_path):

    if zipfile.is_zipfile(file_path):
        with zipfile.ZipFile(file_path) as z:
            z.extractall("output")
        print("ZIP extracted")

    elif tarfile.is_tarfile(file_path):
        with tarfile.open(file_path) as t:
            t.extractall("output")
        print("TAR extracted")

    else:
        print("Unknown file format")

extract_file("/content/archive (1).zip")
```
