```python
#Given t test cases, for each, read a list of numbers and output their product.
t = int(input())
for _ in range(t):
  n = int(input())
  arr = list(map(int, input().split()))
  product=1
  for num in arr:
    product*=num
  print(product)
```

```python
#Read a list of integers and print them separated by "->".
arr = list(map(int, input().split()))
print('->'.join(map(str,arr)))
```

```python
a, b = map(int, input('Enter two numbers:').split())
a, b = b, a

print('After Swapping:')
print('a =',a, 'b =',b)
```

```python
x, y, z = input('Enter three values:').split()
print('x =',x)
print('y =',y)
print('z =',z)
```

```python
num = input('Enter a number:')
print('As String:',num)
print('As Integer:',int(num))
print('As float:',float(num))
```

```python
a = 10
b = 3.14
c = 'Python'
d = True

print(type(a))
print(type(b))
print(type(c))
print(type(d))
```

```python
#Constants vs Variables in Python
PI = 3.14159
radius = int(input('Enter radius:'))

area = PI * (radius ** 2)
print('Area of circle:',area)
```

```python
#Assign Same Value to Multiple Variables

x = y = z = 100
print('x =',x, 'y =',y, 'z =',z)
```

PYTHON OPERATORS

```python
a = 15
b = 4

#Addition
print('Addition:',a+b)

#Subtraction
print('Subtraction:',a-b)

#Multiplication
print('Multiplication:',a*b)

#Division
print('Division:',a/b)

#Floor Division
print('Floor Division:',a//b)

#Modulus
print('Modulus:',a%b)

#Exponential
print('Exponential:',a ** b)
```

```python
a = 13
b = 33

print(a > b)
print(a == b)
print(a < b)
print(a >= b)
print(a <= b)
print(a != b)
```

```python
a = 10
b = 4

print(a & b)
print(a | b)
print(~a)
print(a ^ b)
print(a >> 2)
print(a << 2)
```
