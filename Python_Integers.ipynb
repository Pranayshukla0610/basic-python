```python
#Compute factorial of 1000 and get the number of digits
import math
n = 1000
factorial = math.factorial(n)
num_digits = len(str(factorial))
print("Number of digitd in 1000!:",num_digits)
```

```python
#Bitwise AND for Even/Odd Checking
def is_odd(n):
  return n & 1 == 1

print(is_odd(17))
print(is_odd(23))
```

```python
#Bitwise OR for Flags
READ = 0b001
WRITE = 0b010
EXECUTE = 0b100
user_perm = READ | EXECUTE
print(bin(user_perm))
```

```python
#Left Shift for Multiplication
n = 7
print(n << 3)
```

```python
#Right Shift for Division
n = 100
print(n >> 2)
```
