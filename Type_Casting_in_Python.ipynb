```python
#Convert User Input into Integer List

user_input = "1 2 3 4"
int_list = list(map(int,user_input.split()))
print(int_list)
```

```python
#Float to Integer Precision Loss

num = 98.76
converted = int(num)
print(converted)
```

```python
#String Boolean Trap

value = "False"
print(bool(value))
```

```python
#List of Strings to Float Sum

nums = ["10.5", "12.5", "15.5"]

total = sum(map(float,nums))
print(total)
```

```python
#Tuple to List Modification

t = (1,2,3)
t = list(t)
t.append(4)

t = tuple(t)
print(t)
```

```python
#Dictionary Keys Type Conversion

data = {"1":"A","2":"B"}
new_dict = {int(k):v for k, v in data.items()}
print(new_dict)
```

```python
#Binary String to Decimal

binary = "1010"
decimal = int(binary,2)
print(decimal)
```

```python
#Mixed List Cleaning

data = ["1", 2, "3", 4.0]
cleaned = [int(x) for x in data]
print(cleaned)
```

```python
#String to List of Characters

word = "hello"
char_list = list(word)
print(char_list)
```

```python
#Safe Type Casting with Exception Handling

data = ["10","20","abc","30"]

cleaned = []
for x in data:
  try:
    cleaned.append(int(x))
  except ValueError:
    continue

print(cleaned)
```
