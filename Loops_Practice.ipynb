```python
#Write a program that counts how many times each character appears in a string using loops only.

text = 'mississippi'
freq = {}

for ch in text:
  if ch in freq:
    freq[ch] += 1
  else:
    freq[ch] = 1

print(freq)
```

```python
#Remove Duplicates from List (without set())

arr = [3, 5, 3, 7, 8, 5, 7]
unique = []

for val in arr:
  if val not in unique:
    unique.append(val)

print(unique)
```

```python
#Find Missing Number in a Sequence Using Loops

nums = [1, 2, 4, 5, 6]

left = nums[0]          # smallest number → 1
right = nums[-1]        # largest number → 6

missing = []

for n in range(left, right + 1):
    if n not in nums:
        missing.append(n)

print(missing)
```
