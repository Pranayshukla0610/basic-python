```python
batter = pd.read_csv(r'C:\\Users\\prana\\Documents\\Python data\\batter.csv')
batter.head(5)
```

```python
sample_data = batter.head(100).sample(25,random_state=5)
sample_data
```

```python
plt.scatter(sample_data['avg'],sample_data['strike_rate'])
```

```python
plt.figure(figsize=(18,10))
plt.scatter(sample_data['avg'],sample_data['strike_rate'])
for i in range(sample_data.shape[0]):
    plt.text(sample_data['avg'].values[i],sample_data['strike_rate'].values[i],sample_data['batter'].values[i])
```

```python
plt.figure(figsize=(18,10))
plt.scatter(sample_data['avg'],sample_data['strike_rate'],s=sample_data['runs'])
for i in range(sample_data.shape[0]):
    plt.text(sample_data['avg'].values[i],sample_data['strike_rate'].values[i],sample_data['batter'].values[i])
```

```python
# Players whose strike rate is greater than 130
plt.figure(figsize=(18,10))
plt.scatter(sample_data['avg'],sample_data['strike_rate'],s=sample_data['runs'])

#axis horizontal line
plt.axhline(130,color='red')

for i in range(sample_data.shape[0]):
    plt.text(sample_data['avg'].values[i],sample_data['strike_rate'].values[i],sample_data['batter'].values[i])
```

```python
# Players whose strike rate is greater than 130 and average is more than 30
plt.figure(figsize=(18,10))
plt.scatter(sample_data['avg'],sample_data['strike_rate'],s=sample_data['runs'])

#axis horizontal line
plt.axhline(130,color='red')
plt.axvline(30,color='red')

for i in range(sample_data.shape[0]):
    plt.text(sample_data['avg'].values[i],sample_data['strike_rate'].values[i],sample_data['batter'].values[i])
```

```python
plt.figure(figsize=(18,10))
plt.scatter(sample_data['avg'],sample_data['strike_rate'],s=sample_data['runs'])

#axis horizontal line
plt.axhline(130,color='red')
plt.axhline(140,color='green')
plt.axvline(30,color='red')

for i in range(sample_data.shape[0]):
    plt.text(sample_data['avg'].values[i],sample_data['strike_rate'].values[i],sample_data['batter'].values[i])
```

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(-10,10,100)
y = np.linspace(-10,10,100)
```

```python
fig = plt.figure(figsize=(12,8))
ax = plt.subplot(projection='3d')
p = ax.plot_surface(xx,yy,z, cmap='viridis')
fig.colorbar(p)
```

```python
z = np.sin(xx) + np.cos(yy)
fig = plt.figure(figsize=(12,8))
ax = plt.subplot(projection='3d')
p = ax.plot_surface(xx,yy,z, cmap='viridis')
fig.colorbar(p)
```

```python
z = np.sin(xx) + np.log(yy)
fig = plt.figure(figsize=(12,8))
ax = plt.subplot(projection='3d')
p = ax.plot_surface(xx,yy,z, cmap='viridis')
fig.colorbar(p)
```

```python
z = xx**2+yy**2
fig = plt.figure(figsize=(12,8))
ax = plt.subplot()
p = ax.contour(xx,yy,z, cmap='viridis')
fig.colorbar(p)
```

```python
z = xx**2+yy**2
fig = plt.figure(figsize=(12,8))
ax = plt.subplot()
p = ax.contourf(xx,yy,z, cmap='viridis')
fig.colorbar(p)
```
