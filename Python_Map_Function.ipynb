```python
numbers = [1,2,3,4]
result = map(lambda x: x * 2, numbers)
print(list(result))
```

```python
nums = [1,2,3]
result = []

for n in nums:
  result.append(n * 2)

print(result)
```

```python
nums = [1,2,3,4]
result = list(map(lambda x:x ** 2, nums))
print(result)
```

```python
values = ["1","2","3"]
result = list(map(int, values))
print(result)
```

```python
nums = [1,2,3]
result = list(map(float,nums))
print(result)
```

```python
names = ["alice","bob","charlie"]
result = list(map(str.upper,names))
print(result)
```

```python
nums = [5, 10, 15]
result = list(map(lambda x:x+10,nums))
print(result)
```

```python
a = [1,2,3]
b = [4,5,6]

result = list(map(lambda x,y: x+y,a,b))
print(result)
```

```python
words = ["Python","is","fun"]
lengths = list(map(len,words))
print(words)
```
