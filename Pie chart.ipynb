```python
import matplotlib.pyplot as plt

# Data for Pie Chart
labels = ['Utilized Space', 'Unutilized Space']
pre_utilized = 90000  # 90% of 100,000 square feet
pre_unutilized = 10000  # 10% of 100,000 square feet
post_utilized = 95000  # 95% of 100,000 square feet
post_unutilized = 5000  # 5% of 100,000 square feet

# Pie Chart Data
pre_data = [pre_utilized, pre_unutilized]
post_data = [post_utilized, post_unutilized]

# Pie Chart for Pre-Implementation
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.pie(pre_data, labels=labels, autopct='%1.1f%%', startangle=140, colors=['#4caf50', '#f44336'])
plt.title('Pre-Implementation Space Utilization')

# Pie Chart for Post-Implementation
plt.subplot(1, 2, 2)
plt.pie(post_data, labels=labels, autopct='%1.1f%%', startangle=140, colors=['#4caf50', '#f44336'])
plt.title('Post-Implementation Space Utilization')

plt.show()

# Data for Line Graph
time_periods = ['Pre-Implementation', 'Post-Implementation']
utilized_space = [90000, 95000]
unutilized_space = [10000, 5000]

# Line Graph
plt.figure(figsize=(10, 6))
plt.plot(time_periods, utilized_space, marker='o', label='Utilized Space')
plt.plot(time_periods, unutilized_space, marker='o', label='Unutilized Space')

plt.xlabel('Time Period')
plt.ylabel('Square Feet')
plt.title('Warehouse Space Utilization: Pre vs. Post Implementation')
plt.legend()

plt.show()
```
