```python
#Check even or odd using %
n = 7
if n % 2 == 1:
  print(n,"is odd")
else:
  print(n,"is even")
```

```python
#Check even or odd using bitwise AND
n = 8
if n & 1:
  print(n, "is odd")
else:
  print(n, "is even")
```

```python
#Bitwise OR
a = 0b1010
b = 0b1100

c = a|b
print(bin(c))
```

```python
#Bitwise AND
a = 0b1010
b = 0b1100

c = a & b
print(bin(c))
```

```python
#Bitwise XOR
a = 0b1010
b = 0b1100

c = a^b
print(bin(c))
```

```python
#Left shift (<<)
#n << k multiplies by 2**k
#5 << 2 = 5 * 4 = 20
n = 5
print(n << 2)
```

```python
#Right shift (>>)
#n >> k divides by 2**k (floor division)
#20 >> 2 = 20 // 4 = 5
n = 20
print(n >> 2)
```

```python
#Large Integer Modulo
a = 10**50
b = 12345

c = a%b
print(c)
```

```python
#Swap Two Numbers Using XOR
a = 12
b = 25

a = a ^ b
b = a ^ b
c = a ^ b

print(a,b)
```

```python
#Left Shift for Multiplication
n = 7
k = 3
result = n << k
print(result)
```

```python
#Right Shift for Division
n = 45
k = 2

result = n >> k
print(result)
```
