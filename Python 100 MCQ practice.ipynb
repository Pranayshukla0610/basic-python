```python
✅ Basic OOP Concepts in Python:
Python supports Object-Oriented Programming.
Python allows defining and using classes and objects.

Everything in Python is an object.
Even integers, strings, functions, etc., are objects.

A class is a blueprint for creating objects.
It defines attributes (variables) and behaviors (methods).

An object is an instance of a class.
You can create many objects from one class.

The __init__() method is a constructor.
It runs automatically when an object is created and is used to initialize object attributes.

Attributes can be instance-specific or class-level.

Instance attributes are unique to each object.

Class attributes are shared by all instances.

self refers to the current object.
It is always the first argument in instance methods.

Python supports multiple inheritance.
A class can inherit from more than one base class.

Private attributes use underscores.
self.__data is a private variable (name mangling).

Magic Methods like __str__, __len__, __eq__ add special behavior to your classes.

A class is just a blueprint, template, or design — like an architect's drawing.

An object is the real thing created based on that blueprint — like an actual building.

```
