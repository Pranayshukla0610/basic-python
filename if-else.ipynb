```python
# Write a Python script that keep on asking the user to enter the numbers and add them, and get stops when user enters 0 and show th total

sum = 0

num = int(input('Enter a number (0 to stop):"))

while num != 0:
    sum = sum + num
    num = int(input('Enter another number (0 to stop):'))

print("Total sum:",sum)
```

```python
#Write a python script that keeps on asking the user to deposit money until they reach atleast 10000

goal = 10000
total = 0
while total < goal:
    deposit = float(input('Enter a number'))
    total = total+deposit

print('Goal Reached')
print('Total Balance amount is:',total)
```

```python
#Write a python script, allow a user 3 attempts only to enter the correct password, then lock them out after 3 failed attempts

correct_password = "abc1234"
n = 0
max_n = 3

while n < max_n:
    user_pass = input('Enter your password')
    if user_pass == correct_password:
        print("You have correctly entered the number")
        break
    else:
        print('Incorrect Password')
        n += 1

if n == max_n:
    print('Too many Incorrect Passwords! Access Denied')
```
