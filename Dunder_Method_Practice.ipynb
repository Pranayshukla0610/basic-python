```python
class vehicle:
  def __init__(self,brand):
    self.brand = brand

class Car(vehicle):
  def __init__(self,brand,model):
    super().__init__(brand)
    self.model = model

class ElectricCar(Car):
  def __init__(self,brand,model,battery):
    super().__init__(brand,model)
    self.battery = battery

e = ElectricCar("Tesla","Model S", "100KWH")
print(e.brand, e.model, e.battery)
```

```python
class A:
  def __init__(self):
    print("A")
    super().__init__()

class B:
  def __init__(self):
    print("B")
    super().__init__()

class C(A,B):
  def __init__(self):
    print("C")
    super().__init__()

c = C()
```

```python
class A:
  def __init__(self):
    print("A")

class B:
  def __init__(self):
    print("B")

class C(A,B):
  def __init__(self):
    super().__init__()

c = C()
```

```python
#Passing Arguments using super()
class Parent:
  def __init__(self,name):
    self.name = name

class Child(Parent):
  def __init__(self,name,age):
    super().__init__(name)
    self.age = age

c = Child("Raman",27)
print(c.name,c.age)
```

```python
class Circle:
  def __init__(self,radius):
    self.radius = radius
    self.area = 3.14 * radius * radius

c = Circle(4)
print(c.area)
```

```python
class Employee:
  def __init__(self, name, salary = 50000):
    self.salary = salary
    self.name = name

e1 = Employee("Raman")
e2 = Employee("Sohan",80000)
print(e1.salary, e2.salary)
```
