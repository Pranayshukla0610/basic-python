```python
#High-Precision Average of Large Numbers

numbers = [10**12 + i for i in range(1,6)]

average = sum(numbers)/len(numbers)

print("Numbers:",numbers)
print("Average:",average)
```

```python
#Cyclic Time Conversion Using Modulus

seconds = 987654321

days = seconds // (24 * 3600)
seconds %= (24 * 3600)
hours = seconds // 3600
seconds %= 3600
minutes = seconds // 60
seconds %= 60

print(f"{days} days, {hours} hours, {minutes} minutes, {seconds} seconds")
```

```python
#Newton-Raphson Method for Square Root

def newton_sqrt(n, tolerance=1e-10):
  x = n
  while True:
    root = 0.5 * (x + n / x)
    if abs(root - x) < tolerance:
      return root
    x = root

number = 612
print(f"Square root of {number}: {newton_sqrt(number)}")
```

```python
#Detecting Leap Years Using Arithmetic Logic

def is_leap_year(year):
  return(year % 4 == 0 and year % 100 != 0) or (year % 400 == 0)

years = [1900, 2000, 2024, 2021]
for y in years:
  print(f"{y} is a leap year: {is_leap_year(y)}")
```

```python
#Fast Exponentiation Using Binary Method

def fast_power(base, exponent):
  result = 1
  while exponent > 0:
    if exponent % 2 == 1:
      result *= base
    base *= base
    exponent //= 2
  return result

print("Result:", fast_power(3,13))
```

```python
#Digital Root Using Modulus Arithmetic

def digital_root(n):
  if n == 0:
    return 0
  return 1 + (n - 1)% 9

number = 987654321123456789
print("Digital Root:",digital_root(number))
```

```python
#Simulating Compound Interest with Exponentiation
# Calculate compound interest

def compound_interest(principal, rate, times, years):
    amount = principal * (1 + rate / times) ** (times * years)
    return amount

principal = 10000
rate = 0.08
times = 4  # Quarterly compounding
years = 5

print("Final Amount:", compound_interest(principal, rate, times, years))
```

```python
#Custom Implementation of Floor Division

def custom_floor_division(a, b):
  quotient = int(a / b)
  if ( a / b != 0) and ((a < 0) != (b < 0)):
    quotient -= 1
  return quotient

print(custom_floor_division(7, 3))
print(custom_floor_division(-7, 3))
print(custom_floor_division(7, -3))
print(custom_floor_division(-7, -3))
```
