```python
# LIST = List is a data type where you can store multiple items under 1 name. More technically list acts like a dynamic arrays which meands you can add more items on the fly.
# List are heterogeneous, while array are homogeneous
# List occupy more space and time in comparison to array
# Referrential array
# Dynamic Array
# id
# Heterogeneous array vs homogeneous array
```

```python
# CHARACTERSTICS OF LIST
# List are ordered
# List are changeable/Mutable
# List is heterogeneous
# In List, items can be duplicate
# List are dynamic
# List can be nested
# Items can be accessed
# List can contain any type of object
```

```python
# Empty list
print([])
```

```python
# 1D List
print([1,2,3,4,5])
```

```python
# 2D List
print([1,2,3,[4,5]])   # Its a Heterogeneous list
```

```python
# 3D List
print([[[1,2]],[[3,4]],[[5,6]],[[7,8]]])  # Its a Homogeneous list
```

```python
# Heterogeneous list
print([1,True, 5.6,5+6j,'Hello'])
```

```python
# Type conversion
print(list('hello'))
```

```python
# Indexing
L = [1,2,3,4,5]
print(L[0])
print(L[-1])
```

```python
L = [1,2,3,[4,5]]
print(L[3])
print(L[3][1])
print(L[-1][-2])
```

```python
L = [[[1,2],[3,4]],[[5,6],[7,8]]]
print(L[0])
print(L[1])
print(L[1][0])
print(L[1])
print(L[1][0][0])
print(L[0][1][0])
```

```python
# Slicing
L = [1,2,3,4,5,6]
print(L[0:3])
print(L[-3:])
print(L[0::2]) # Alternate
print(L[-5:-2:2])
print(L[::-1])
```

```python
# Append
L = [1,2,3,4,5]
L.append(True)
print(L)
```

```python
L = [1,2,3,4,5]
L.append([6,7,8])
print(L)
```

```python
# Extend
L = [1,2,3,4,5]
L.extend([6,7,8])
print(L)
```

```python
L = [1,2,3,4,5]
L.extend('delhi')
print(L)
```

```python
# Insert
L = [1,2,3,4,5]
L.insert(1,100) # 1 is index position where 100 is added
print(L)
```

```python
# Editing wit indexing
L = [1,2,3,4,5]
L[-1] = 500
print(L)
# Editing wit slicing
L[1:4] = [200,300,400]
print(L)
```

```python
# Deletion with indexing
L = [1,2,3,4,5]
del L[-1]
print(L)

# Deletion with slicingabs
del L[1:3]
print(L)
```

```python
# Remove
L = [1,2,3,4,5]
L.remove(5)
print(L)
```

```python
# POP
L = [1,2,3,4,5]
L.pop(0)
print(L)
```

```python
# Clear
L = [1,2,3,4,5]
L.clear()
print(L)
```

```python
# ARITHMETIC
# Addition
L1 = [1,2,3,4]
L2 = [5,6,7,8]

# Concatenate/Merge
print(L1+L2)
```

```python
print(L1*3)
```

```python
L1 = [1,2,3,4,5]
L2 = [1,2,3,4,[5,6]]
print(5 in L1)
```

```python
L1 = [1,2,3,4,5]
L2 = [1,2,3,4,[5,6]]
print(5 not in L1)
print(5 in L2)
```

```python
# LOOPS
L1 = [1,2,3,4,5]
L2 = [1,2,3,4,[5,6]]

for i in L1:
    print(i)
```

```python
L1 = [1,2,3,4,5]
L2 = [1,2,3,4,[5,6]]

for i in L2:
    print(i)
```

```python
# len/min/max/sorted
L = [2,1,5,7,0]
print(len(L))
print(min(L))
```
