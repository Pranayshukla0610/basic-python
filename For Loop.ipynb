```python
#Given a dictionary of items and their quantities, and fixed unit price. Calculate total cost per item and total bill

order = {"Apples":2, "Banana":6, "Oranges":3}
unit_price = 20
total_bill = 0

for i in order:
    print(i,order[i]*unit_price)
```

```python
order = {"Apples":2, "Banana":6, "Oranges":3}
unit_price = 20
total_bill = 0

for item, quantity in order.items():
    cost = quantity * unit_price
    print(f"{item}:{quantity} * {unit_price} = {cost}")

    total_bill = total_bill + cost
print(f"Total Bill: {total_bill}")
```

```python
#Check a warehouse inventory list if quantity < 10, mark the item for reorder

inventory = {"laptops":25, "keyboards":8, "Monitors":15, "Mouse":5}
req_qty= 10

for item, qty in inventory.items():
    if qty < 10:
        print(f"{item} --> reorder")
    else:
        print(f"{item} --> sufficient")
```

```python
#Python script to check the password strength for a list of passwords. Flag the password weak, if it has less than 8 characters

passwords = ['abc123', 'samraat567', 'hello', 'India1947', 'admin@987654321']

for i in passwords:
    if len(i)<8:
        print(f"{i} --> weak password")
    else:
        print(f"{i} --> strong password")
```
