```python
#map() + lambda
#Purpose: Apply function to every item in iterable.
nums = ['1','2','3']
squares = list(map(lambda x : int(x) ** 2,nums))
print(squares)
```

```python
#Filter items based on condition.
nums = [5,8,13,19]
evens = list(filter(lambda x : x % 2 == 0,nums))
print(evens)
```

```python
#reduce() from functools
#Purpose: Applies rolling computation to elements.
from functools import reduce
nums = [1,2,3,4]
produce = reduce(lambda x, y : x * y, nums)
print(produce)
```

```python
#enumerate()
#Purpose: Add index to iterable.
colors = ['red','green','blue']
for i, color in enumerate(colors):
    print(i,color)


#In the for loop, i gets the index (0, 1, 2...), and color gets the item from the list.

#The print(i, color) line prints both the index and the color.
```

```python
#all() and any()
#Purpose:

#all() returns True if all elements are true.

#any() returns True if at least one is true.

data = [True,True,False]
print(all(data))
print(any(data))
```

```python
#isinstance() and issubclass()
#Purpose: Type checking.

print(isinstance(5,int))
```

```python
class A: pass
class B(A): pass

print(issubclass(B,A))
```

```python
#vars() and dir()
#Purpose:

#vars(): Return __dict__ of an object.

#dir(): Return valid attributes.

class Text:
    def __init__(self):
        self.a = 10
        self.b = 20

t = Text()
print(vars(t))
print(dir(t))
```

```python
#Import a module using a string and call a function
module_name = "math"
function_name = "sqrt"
mod = __import__(module_name)
result = getattr(mod, function_name)(16)
print(result)




#What is __import__() in Simple Words?
#__import__() is a built-in Python function.

#It is used when you want to import a module while your program is running, not at the beginning.

#This is called dynamic importing.

#ðŸ§  Why/When to Use It?
#Use it when:

#You donâ€™t know the module name until the program runs.

#You want to load modules based on user input or config files.

#You are building a plugin system.

#You want to avoid loading all modules unless needed (save memory).
```

```python
module_name = 'math'
function_name = 'factorial'

args = 5

mod = __import__(module_name)
result = getattr(mod,function_name)(args)
print(result)
```

```python
modules = ['math','random','datetime']
for mod_name in modules:
    mod = __import__(mod_name)
    print(f"Loaded module: {mod.__name__}")
```
