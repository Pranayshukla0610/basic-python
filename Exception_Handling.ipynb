```python
try:
  num = int(input("Enter a number: "))
  print(num)
except ValueError:
  print("Please Enter a valid number")
```

```python
#How do you handle multiple exceptions?
try:
  a = int(input("Enter a number:"))
  print(10/a)
except ZeroDivisionError:
  print("Cannot Divide by zero")
except ValueError:
  print("Invalid Output")
```

```python
#What is the purpose of the else block?
try:
  num = int(input("Enter a number:"))
  result = 10/num
except ZeroDivisionError:
  print("Division by zero error")
else:
  print("Result:",result)
```

```python
numbers = [10, 20, 30]
try:
  print(numbers[5])
except IndexError:
  print("Index out of range")
```

```python
#Can exception handling be used inside functions?
def divide(a,b):
  try:
    return a/b
  except ZeroDivisionError:
    return "Division by zero not allowed"
print(divide(10,0))
```

```python
#How do you create and handle a custom exception?

def check_score(score):
  if score < 0 or score > 100:
    raise ValueError("Score must be between 0 and 100")
  return "Valid score"

try:
  print(check_score(150))
except ValueError as e:
  print(e)
```

```python
#How do exceptions behave inside loops?
numbers = [10, 5, 0, 2]

for n in numbers:
    try:
        print(10 / n)
    except ZeroDivisionError:
        print("Skipped division by zero")
```

```python
#Role of __exit__ in exception handling?
class Demo:
    def __enter__(self):
        print("Enter block")
        return self

    def __exit__(self, exc_type, exc_value, traceback):
        print("Exit block")
        print("Exception:", exc_type)
        return True

with Demo():
    print(10 / 0)
```
