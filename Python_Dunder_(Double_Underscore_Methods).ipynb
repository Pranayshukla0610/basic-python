```python
#Difference between __new__ and __init__

class A:
  def __new__(cls):
    print("Creating Instance")
    return super().__new__(cls)

  def __init__(self):
    print("Initializing Instance")

a = A()
```

```python
#Use of super()

class Parent:
  def __init__(self):
    print("Parent init")

class Child(Parent):
  def __init__(self):
    super().__init__()
    print("Child init")

c = Child()
```

```python
#__str__ vs __repr__

class A:
  def __str__(self):
    return "User View"

  def __repr__(self):
    return "Developer View"

a = A()
print(a)
```

```python
#Operator Overloading (__add__)
class Point:
    def __init__(self, x):
        self.x = x

    def __add__(self, other):
        return Point(self.x + other.x)

p1 = Point(10)
p2 = Point(20)

print((p1 + p2).x)
```

```python
#Custom Length (__len__)

class MyList:
  def __init__(self,data):
    self.data = data

  def __len__(self):
    return len(self.data) * 2

obj = MyList([1,2,3])
print(len(obj))
```

```python
#Indexing (__getitem__)

class MyList:
  def __init__(self,data):
    self.data = data

  def __getitem__(self,index):
    return self.data[index] * 10

obj = MyList([1,2,3])
print(obj[1])
```
