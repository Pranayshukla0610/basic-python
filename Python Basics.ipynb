```python
# Multiple Inheritance in Python
class Parent1():
    def assign_string_one(self,str1):
        self.str1 = str1
        
    def show_string_one(self):
        return self.str1   
```

```python
class Parent2():
    def assign_string_two(self,str2):
        self.str2 = str2
    def show_string_two(self):
        return self.str2
```

```python
class Derived(Parent1, Parent2):
    def assign_string_three(self, str3):
        self.str3 = str3
    def show_string_three(self):
        return self.str3
    
```

```python
d1 = Derived()
```

```python
d1.assign_string_one("one")
d1.assign_string_two("two")
d1.assign_string_three("three")
```

```python
d1.show_string_one()
```

```python
d1.show_string_two()
```

```python
d1.show_string_three()
```

```python
# Dynamic Typing
In simple terms, dynamic typing in Python means you dont have to specify the type of a variable (like number, text, etc.) when you create it. 
Python figures out the type on its own based on what you assign to it. Python is a language that supports Dynamic Typing
```

```python
# Dynamic Typing
x = 5        # x is a number (integer)
x = "Hi"     # Now x is a text (string)
print(x)
```

```python
# Dynamic Binding
In simple terms, dynamic binding in Python means that the method or function that gets called is decided at runtime, rather than in advance. 
This allows Python to choose the appropriate method based on the type of object being used at that moment.
```

```python
class Animal:
    def speak(self):
        print("Some sound")

class Dog(Animal):
    def speak(self):
        print("Woof!")

class Cat(Animal):
    def speak(self):
        print("Meow!")

def make_animal_speak(animal):
    animal.speak()

dog = Dog()
cat = Cat()

make_animal_speak(dog)  # Outputs: Woof!
make_animal_speak(cat)  # Outputs: Meow!
```

```python
# Static Typing
In simple terms, static typing means that the type of a variable is fixed and has to be declared when you create it, and it cant change later on. However, Python itself is dynamically typed, so it doesnt require you to declare types for variables. 
But Python has a way to add static typing through type hints. C, c++, JAVA supports static typing
```

```python
def add(a: int, b: int) -> int:
     return a + b
```

```python
# Static Binding
In simple terms, static binding means that the method or function to be called is decided at compile time (before the program runs). 
Since Python is dynamically typed and interpreted, it primarily uses dynamic binding, but some behaviors in Python can resemble static binding.
C, c++, JAVA supports Static Binding
```

```python
class Animal:
    def eat(self):
        print("Animal is eating")

animal = Animal()
animal.eat()  # Outputs: Animal is eating
```

```python
# Elif and Nested if-else

email = input('enter email')
password = input('enter password')

if email == 'nitish.campusx@gmail.com' and password == '1234':
    print('Welcome')
elif email == 'nitish.campusx@gmail.com' and password != '1234':
    # tell the user
    print('Incorrect password')
    password = input('Enter password again')
    if password == '1234':
        print('Welcome finally')
    else:
        print('beta tumse na ho payega!')
else:
    print('Not Correct')
```

```python
# Menu driven calculator
fnum = int(input('enter the 1st number'))
snum = int(input('enter the 2nd number'))

op = input('Enter the operation')

if op == '+':
    print(fnum + snum)
elif op == '-':
    print(fnum - snum)
elif op == '*':
    print(fnum * snum)
else:
    op == '/'
    print(fnum / snum)
```

```python
# While loop with else
number = int(input('Enter the number'))

i = 1 # i stand for iteration
while i<11:
    print(number * i)
    i += 1
```

```python
# While loop with else
number = int(input('Enter the number'))

i = 1 # i stand for iteration
while i<11:
    print(number, '*',i,'=', number * i)
    i += 1
```

```python
# while loop with else

X = 1

while X < 3:
    print(X)
    X += 1
else:
    print('limit crossed')
```

```python
# Sequence sum 1/1! + 2/2! + 3/3! + .....
# Find sum till nth term
n = int(input('Enter n'))

result = 0
fact = 1
for i in range(1,n+1):
    fact = fact * i
    result = result + i/fact
print(result)
    
    
```
