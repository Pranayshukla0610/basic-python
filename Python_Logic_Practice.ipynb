```python
#Check if array is a palindrome

#Input: [1,2,3,2,1]
#Output: True

def is_palindrome(arr):
  left = 0
  right = len(arr) - 1

  while left < right:
    if arr[left] != arr[right]:
      return False
    left += 1
    right -+1
  return True

arr = [1,2,3,2,1]
print(is_palindrome(arr))
```

```python
#Find two numbers that are sum to target
#Array: [1, 2, 3, 4, 6]
#Target: 6

def two_sum(arr, target):
  left = 0
  right = len(arr) - 1
  result = 0

  while left < right:
    current_sum = arr[left] + arr[right]
    if current_sum == target:
      return[left,right]

    elif current_sum < target:
      left += 1
    else:
      right -= 1

arr = [1,2,3,4,6]
print(two_sum([1,2,3,4,6],6))
```

```python
#Remove duplicates from sorted arrays
#Input = [1,1,2,2,3]
#Output = [1,2,3]

def remove_duplicates(arr):
  left = 0
  for right in range(1, len(arr)):
    if arr[right] != arr[left]:
      left +=1
      arr[left] = arr[right]
  return left+1

arr = [1,1,2,2,3]
len = remove_duplicates(arr)
print(arr[:len])
```

```python
#Reverse an array using two pointers
#Input:[1,2,3,4]
#Output: [4,3,2,1]

# Fix if len was accidentally overwritten
try:
    len([1, 2, 3])            # test if len is still callable
except TypeError:
    del len                   # restore the built-in len if overwritten


def reverse_arr(arr):
    left = 0
    right = len(arr) - 1

    while left < right:
        arr[left], arr[right] = arr[right], arr[left]
        left += 1
        right -= 1
    return arr


print(reverse_arr([1, 2, 3, 4]))

```

```python
#Move all zeroes to end
#Input:[0,1,0,3,12]
#Output:[1,3,12,0,0]

def move_zeroes(arr):
  l = 0
  for r in range(len(arr)):
    if arr[r] != 0:
      arr[l], arr[r] = arr[r], arr[l]
      l += 1
  return arr

print(move_zeroes([0,1,0,3,12]))
```

```python
#Check if string is palindrome
#Input: 'level'
#Output: True

def is_palindrome(s):
  l = 0
  r = len(s) - 1

  while l < r:
    if s[l] != s[r]:
      return False
    l += 1
    r -= 1
  return True

print(is_palindrome('level'))
```

```python
#Reverse a list using loop
#Input = [10,20,30,40,50]

def reverse_list(arr):
  l = 0
  r = len(arr) - 1

  while l < r:
    arr[l], arr[r] = arr[r],arr[l]
    l += 1
    r -= 1

  return arr

arr = [10,20,30,40,50]
result = reverse_list(arr)
print(result)
```
