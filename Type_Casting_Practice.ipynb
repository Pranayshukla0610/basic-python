```python
#Implicit vs Explicit Casting Trap

a = 5
b = 2.0

print(a+b, type(a+b))
```

```python
#Boolean Arithmetic

print(True + True * False)
```

```python
#Complex Number Casting

num = "3+4j"
c = complex(num)
print(c)
```

```python
#Float Precision Bug

x = 0.1 + 0.2
print(x == 0.3)
```

```python
round(x,2)
```

```python
#Nested Casting

data = ["1","2.5","3"]
result = [int(float(x)) for x in data]
print(result)
```

```python
#Dictionary from Two Lists

keys = ["1","2","3"]
values = ["10","20","30"]

d = dict(zip(map(int,keys),map(int,values)))
print(d)
```

```python
#Set Casting Removes Duplicates

nums = ["1","2","2","3"]
unique = set(map(int,nums))
print(unique)
```

```python
#Bytes to String Decoding

b = b'hello'
print(str(b))
```

```python
#Casting with Default Values

data = ["10", "", "30"]
result = [int(x) if x else 0 for x in data]
print(result)
```

```python
#Pandas Type Casting

import pandas as pd
df = pd.DataFrame({"price":["10","20","30"]})
df['price'] = df['price'].astype(int)
print(df)
```

```python
#Lambda + Casting

nums = ["1","2","3"]
squared = list(map(lambda x: int(x)**2,nums))
print(squared)
```
