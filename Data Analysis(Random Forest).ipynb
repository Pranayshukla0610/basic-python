```python
#Generating Tags/Features from free text available
import re
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
```

```python
print(data_model.columns)
```

```python
def clean_text(text):
    text = text.lower()
    text = re.sub(r'[^\w\s]', '', text)  # Remove punctuation
    text = re.sub(r'\d+', '', text)      # Remove numbers
    tokens = text.split()
    tokens = [word for word in tokens if word not in stopwords.words('english')]
    return " ".join(tokens)
```

```python
data['CLEANED_CUSTOMER'] = data['CUSTOMER_VERBATIM'].apply(clean_text)
data['CLEANED_CORRECTION'] = data['CORRECTION_VERBATIM'].apply(clean_text)
```

```python
data
```

```python
from collections import Counter

def get_top_words(series, n=10):
    all_words = " ".join(series.dropna()).split()
    return Counter(all_words).most_common(n)

print("Top Customer Complaint Tags:")
print(get_top_words(data['CLEANED_CUSTOMER'], n=10))

print("\nTop Correction Tags:")
print(get_top_words(data['CLEANED_CORRECTION'], n=10))
```
