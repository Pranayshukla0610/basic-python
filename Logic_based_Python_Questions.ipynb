```python
#check palindrome(String)
def is_palindrome(s):
  return s == s[::-1]

print(is_palindrome('madam'))
```

```python
#check palindrome(Number)

def is_palindrome(n):
  original = n
  rev = 0

  while n > 0:
    rev = rev * 10 + n % 10
    n //= 10

  return original == rev

print(is_palindrome(121))
```

```python
#Factorial

def factorial(n):
  if n == 0:
    return 1
  return n * factorial(n - 1)

print(factorial(5))
```

```python
#Fibonacci series

def fibonacci(n):
  a, b = 0, 1
  for _ in range(n):
    print(a, end=" ")
    a, b = b, a +b

fibonacci(7)
```

```python
#Prime Number Check

def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n**0.5)+1):
        if n % i == 0:
            return False
    return True

print(is_prime(7))
```
